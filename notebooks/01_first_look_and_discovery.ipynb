## 2. Discover a real truth yourself

The safest, most satisfying early wins are **things you find in the data** — un-leakable, and they *are* the core lesson. Run the three cells below. Each is ~10 lines of pandas and each overturns a common SEO belief.

Every number is **computed live from the shipped CSV** — nothing is hardcoded.

```python
import pandas as pd, numpy as np
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
print(df.shape[0], "rows,", df.shape[1], "columns")
df.head(3)
```

### Discovery A — "High search volume means more traffic." Does it?

```python
corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d: {corr:.3f}")
print("Near zero -> keyword search volume barely predicts the traffic a page actually gets.")
```

### Discovery B — the CTR cliff by position
Click-through rate is not flat: it collapses as you move down the results.

```python
visible = df[df["impressions_90d"] >= 100]
ctr_by_pos = visible.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print(ctr_by_pos.round(4).to_string())
ctr_by_pos.plot(kind="bar", title="Mean CTR by position tier (impressions >= 100)", ylabel="CTR");
```

### Discovery C — is longer content the lever?
Compare word count for **declining** vs **growing** pages.

```python
wc = df.groupby("trend_direction")["word_count"].median()
print(wc.round(0).to_string())
print("\n'down' vs 'up' pages have almost the same median word count -> length is not the lever.")
```

```python
# Your discovery here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the starter dataset
df = pd.read_csv('data/raw/content_refresh_anonymized.csv')

# 2. Calculate the baseline correlation between position and CTR
correlation = df['avg_position'].corr(df['ctr'])
print(f"Overall Pearson correlation between avg_position and CTR: {correlation:.4f}")

# 3. Create a clean aggregation to observe the trend across position tiers
# We round positions to get clear, distinct ranks (e.g., Position 1, 2, 3...)
df['rank_tier'] = df['avg_position'].round()
position_summary = df[df['rank_tier'] <= 10].groupby('rank_tier')['ctr'].mean().reset_index()

print("\nAverage CTR by Position Tier (Top 10):")
print(position_summary.to_string(index=False))

# 4. Generate a clean visual anchor for your GitHub portfolio
plt.figure(figsize=(8, 4))
sns.lineplot(data=position_summary, x='rank_tier', y='ctr', marker='o', color='purple')
plt.title('Observed Relationship: Average Position vs. CTR')
plt.xlabel('Rounded Rank Position')
plt.ylabel('Mean CTR')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

Week 1 Discovery: Statistical Audit of Rank Position vs. Click-Through Rate (CTR)Objective:To test the industry assumption regarding how a webpage's average search engine ranking position correlates with its actual user click-through rate (CTR).Methodology:Utilizing the anonymized historical search data sample, a Pearson correlation coefficient was calculated between avg_position and ctr. Furthermore, positions were binned into rounded rank tiers (1 through 10) to observe directional trends in mean user engagement.Observations & Directional Insights:Negative Correlation: The data reveals a distinct negative correlation between the metrics. As numerical rank position increases (meaning the page drops lower on the search results page), the click-through rate demonstrably decays.Non-Linear Decay: The directional trend is non-linear. The drop in CTR from position 1 to position 3 is steep, after which the curve flattens significantly toward position 10. This is consistent with a power-law distribution behavior often observed in human search interaction.Boundaries: These patterns reflect localized behavioral tendencies within this specific anonymized dataset sample. They serve as directional indicators for feature selection and cannot be extrapolated as a definitive model of external search ranking algorithms.
